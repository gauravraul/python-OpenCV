```python
# We are using numpy to display image as 3 dimensional array of numbers
import numpy as np
```

```python
# matplotlib is usually used for plotting graphs
# but we are using it to display images inline.
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# using PIL to just load the image
from PIL import Image
```

```python
pic = Image.open('Images/cats.jpg')
```

```python
pic
```

```python
#Now check the type of the pic
type(pic)
```

```python
# It is a Jpeg Image File type.
# But to perform Operations or apply Machine Learning or Deep Learning Algorithms we need to convert it
# into raw 3 dimensional Array of numbers

pic_arr = np.asarray(pic)
```

```python
# now again check the type
type(pic_arr)
```

```python
# now it is an n- dimensinal array
# lets check its dimensions / shape

pic_arr.shape
```

```python
# the shape says that 
# its resolution is 2160 x 2880 
#(i,e 2160 pixels in heigt and 2880 pixels in width)
# also it has 3 color channel RED, GREEN ,BLUE

#now lets convert this array to image
plt.imshow(pic_arr)
```

```python
# Now the major confusion here is about the 3 channels
# how are they responsible for creating this image?

# All visible spectrum of colors can be made using 3 primary colors Red, Green Blue.
# ths image above is also made by overlapping values of these 3 colors

#we can just check what  'pic_arr' actually is

pic_arr
```

```python
# so we can see that there are lot of values
# the values are nothing but color intensity of pixels in all the 3 channels
#  minimum : 0 means no color i.e Black
# maximum : 255 means White

#so by making combinations of the intensity of color of each pixel we can make images.

# Now lets see how does the image look when we remove any of the 3 primary colors :
# just make all the values of the color channel to 0

#removing Red

#make a copy of original
pic_no_red = pic_arr.copy()

# index of RGB channel is al follows:
# Red = 0, Green = 1 , Blue =2

# making all the rows and columns of red channel (index = 0) as 0
pic_no_red[:,:,0]=0

plt.imshow(pic_no_red)
```

```python
# The above image has no trace of red in it, the above image is made using only combination of Blue and Green channels

#removing Green
pic_no_green = pic_arr.copy()

# making all the rows and columns of green channel (index = 1) as 0
pic_no_green[:,:,1]=0

plt.imshow(pic_no_green)
```

```python
# The above image has no trace of green in it, the above image is made using only combination of Red and Blue channels

#removing Blue
pic_no_blue = pic_arr.copy()

# making all the rows and columns of blue channel (index = 2) as 0
pic_no_blue[:,:,2]=0

plt.imshow(pic_no_blue)
```

```python
# The above image has no trace of Blue in it, the above image is made using only combination of Red nd Green channels
```

### Conclusion :

We can see now how an image is actually made up.
